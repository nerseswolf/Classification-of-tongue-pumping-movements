#                                                DATA VISUALISATION

A scatter plot is the graphical representation of observed value pairs of two statistical characteristics. These pairs of values are entered in a Cartesian coordinate system, resulting in a point cloud. The points can be represented by various small symbols. 

It is hoped that the pattern of points in the scatter plot will provide information about the dependency structure of the two features represented by coordinates.

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets

featuresTrue = pd.read_excel('pumpFeatures.xlsx', sheet_name = 'FeaturesTrue')
featuresFalse = pd.read_excel('pumpFeatures.xlsx', sheet_name = 'FeaturesFalse')

subjectNr = 1 # Selecting test persons
feature1 = 'SquaredSum' # select feature 1
feature2 = 'Steepness'  # select feature 2

# select data
f1True = featuresTrue[feature1].iloc[np.where(featuresTrue.Subject == subjectNr)].values
f1False = featuresFalse[feature1].iloc[np.where(featuresFalse.Subject == subjectNr)].values
f2True = featuresTrue[feature2].iloc[np.where(featuresTrue.Subject == subjectNr)].values
f2False = featuresFalse[feature2].iloc[np.where(featuresFalse.Subject == subjectNr)].values

# plot data
plt.scatter(f1True, f2True, marker='.', color = 'b' )
plt.scatter(f1False, f2False, marker='x', color = 'k' )
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.title('Scatterplot')
plt.legend(['true', 'false'])

# limit axes
xMax1 = np.percentile(f1True, 95)
xMax2 = np.percentile(f1True, 95)
xMax = xMax1
if xMax1>xMax2:
    xMax = xMax2

yMax1 = np.percentile(f2True, 95)
yMax2 = np.percentile(f2True, 95)
yMax = yMax1
if yMax1>yMax2:
    yMax = yMax2

plt.xlim([0, xMax])
plt.ylim([0, yMax])
plt.show()
```

# 2. Example from the scikit-learn documentation
      -> 3 different flower types

```python

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
Y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()

# plot the training points
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```
